```python
import pandas as pd
```

```python
df = pd.read_excel('Pucha Wholesaler LLC_Purchases by Vendor Detail-CARLOS.xlsx','Hoja1')
```

```python
df
```

```python
df['Transaction date'] = pd.to_datetime(df['Transaction date'])

# Crear un diccionario de DataFrames
ordenes_por_proveedor = {
    (vendor, fecha): grupo.reset_index(drop=True)
    for (vendor, fecha), grupo in df.groupby(['Vendor', 'Transaction date'])
}
```

```python
with pd.ExcelWriter('ordenes_compra.xlsx', engine='xlsxwriter') as writer:
    for i, ((vendor, fecha), grupo) in enumerate(ordenes_por_proveedor.items(), start=1):
        sheet_name = f"OC_{i}_{fecha.date()}"
        grupo.to_excel(writer, sheet_name=sheet_name, index=False)
```
